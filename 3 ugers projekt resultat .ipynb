```python
import requests
import PyPDF2
import spacy
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from scipy.stats import norm

# Load spaCy model for tokenization
nlp = spacy.load("en_core_web_sm")

# Function to extract text from a PDF URL
def extract_pdf_text(url):
    response = requests.get(url)
    if response.status_code == 200:
        with open('temp.pdf', 'wb') as f:
            f.write(response.content)

        with open('temp.pdf', 'rb') as f:
            pdf_reader = PyPDF2.PdfReader(f)
            text = ""
            for page in pdf_reader.pages:
                text += page.extract_text()
            return text
    return ""

# Function to preprocess text: Tokenization using spaCy and chunking
def preprocess_text(text, chunk_size):
    doc = nlp(text)
    sentences = [sent.text for sent in doc.sents]
    chunks = [' '.join(sentences[i:i + chunk_size]) for i in range(0, len(sentences), chunk_size)]
    return chunks

# Function to calculate confidence interval
def calculate_confidence_interval(proportion, n, confidence=0.95):
    z = norm.ppf((1 + confidence) / 2)
    se = np.sqrt((proportion * (1 - proportion)) / n)
    lower_bound = proportion - z * se
    upper_bound = proportion + z * se
    return lower_bound, upper_bound

# Function to evaluate a single model configuration
def evaluate_model(data, chunk_size=3, max_features=1500, test_size=0.35):
    classifier = RandomForestClassifier(n_estimators=100, random_state=42)

    # Split the dataset
    train_data, test_data = train_test_split(list(data.items()), test_size=test_size, random_state=42)

    # Prepare training data
    train_texts, train_labels = [], []
    for url, stock_change in train_data:
        text = extract_pdf_text(url)
        if text:
            chunks = preprocess_text(text, chunk_size)
            train_texts.extend(chunks)
            train_labels.extend([1 if stock_change > 0 else 0] * len(chunks))

    # Prepare testing data
    test_reports, test_labels = [], []
    for url, stock_change in test_data:
        text = extract_pdf_text(url)
        if text:
            chunks = preprocess_text(text, chunk_size)
            test_reports.append((chunks, stock_change))
            test_labels.append(1 if stock_change > 0 else 0)

    # Create and train the model
    pipeline = make_pipeline(
        TfidfVectorizer(max_features=max_features),
        classifier
    )
    pipeline.fit(train_texts, train_labels)

    # Evaluate the model
    correct_predictions = 0
    for chunks, true_label in test_reports:
        chunk_predictions = pipeline.predict(chunks)
        positive_votes = sum(chunk_predictions)
        negative_votes = len(chunk_predictions) - positive_votes
        final_prediction = 1 if positive_votes >= negative_votes else 0
        if final_prediction == true_label:
            correct_predictions += 1

    # Calculate accuracy and confidence interval
    accuracy = correct_predictions / len(test_reports)
    lower_bound, upper_bound = calculate_confidence_interval(accuracy, len(test_reports))

    # Print the results
    print(f"Model Evaluation Complete!")
    print(f"Chunk Size: {chunk_size}")
    print(f"Max Features: {max_features}")
    print(f"Test Size: {test_size}")
    print(f"Accuracy: {accuracy * 100:.2f}%")
    print(f"95% CI: ({lower_bound * 100:.2f}%, {upper_bound * 100:.2f}%)")

 # Evaluate the model with the specified parameters
evaluate_model(data)
```
